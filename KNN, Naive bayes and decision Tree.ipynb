```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('/content/Global_Pollution_Analysis.csv')
df.head()
```

```python
df=df.dropna(how='all')
df.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
df["Industrial_Waste (in tons)"]=scaler.fit_transform(df[["Industrial_Waste (in tons)"]])
df["CO2_Emissions (in MT)"]=scaler.fit_transform(df[["CO2_Emissions (in MT)"]])
pollution_indices = ['Air_Pollution_Index', 'Water_Pollution_Index', 'Soil_Pollution_Index']
df[pollution_indices] = scaler.fit_transform(df[pollution_indices])
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df["Country"]=le.fit_transform(df["Country"])
df["Year"]=le.fit_transform(df["Year"])
df.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, classification_report
columns=["Air_Pollution_Index","Water_Pollution_Index",'Soil_Pollution_Index','Industrial_Waste (in tons)']
y=df["CO2_Emissions (in MT)"]
bins = [y.min() - 1, y.quantile(0.33), y.quantile(0.66), y.max() + 1]
labels = ['Low', 'Medium', 'High']
df['Pollution_Severity'] = pd.cut(df['CO2_Emissions (in MT)'], bins=bins, labels=labels)
x=df[columns]
y=df["Pollution_Severity"]
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
gnb = MultinomialNB()
gnb.fit(x_train, y_train)
y_pred = gnb.predict(x_test)
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy (exact matches): {accuracy}")
classification_report = classification_report(y_test, y_pred)
print("Classification Report:")
print(classification_report)
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
print("\nConfusion Matrix:")
print(cm)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=4)
knn.fit(x_train,y_train)
y_pred=knn.predict(x_test)
accuracy=accuracy_score(y_test,y_pred)
print(f"Accuracy (exact matches): {accuracy}")
# classification_report_str=classification_report(y_test,y_pred)
# print("Classification Report:")
# print(classification_report_str)
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
print("\nConfusion Matrix:")
print(cm)
```

```python
from sklearn.tree import DecisionTreeClassifier
dtc=DecisionTreeClassifier(max_depth=4,min_samples_split=50)
dtc.fit(x_train,y_train)
y_pred=dtc.predict(x_test)
accuracy=accuracy_score(y_test,y_pred)
print(f"Accuracy (exact matches): {accuracy}")
cm=confusion_matrix(y_test,y_pred)
print("\nConfusion Matrix:")
print(cm)
```

This analysis explored a dataset on global pollution, focusing on air, water, and soil pollution indices, industrial waste, and CO2 emissions.

**Data Preprocessing:**
The data was loaded and missing values were handled by dropping rows with all missing values. The 'Country' and 'Year' columns were label encoded for use in the models.

**Pollution Severity Classification:**
To classify pollution severity, the 'CO2_Emissions (in MT)' was discretized into three categories: 'Low', 'Medium', and 'High' based on quantile values.

**Model Evaluation:**
